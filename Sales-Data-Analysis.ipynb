```python
data['Revenue']=data['Price'] * data['Quantity']
```

```python
date_revenue = data.groupby('Date')['Revenue'].sum().reset_index()
```

```python
date_revenue = date_revenue.sort_values('Date')
```

```python
date_revenue.plot(
    x='Date',
    y='Revenue',
    linewidth=2,
    figsize=(9,4)
)

plt.title("Date Wise Revenue")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```
